```python
Fare_min = df.Fare.min()
Fare_max = df.Fare.max()
Age_min = df.Age.min()
Age_max = df.Age.max()
df.agg({
    'Fare': lambda x: (x-Fare_min)/(Fare_max-Fare_min),
    'Age':  lambda x: (x-Age_min)/(Age_max-Age_min)
})
```

```python
df_nan_fare = df.loc[(df.Fare.isna())]
df_nan_fare
```
