# Activity 4.05: Linear Regression for Animal Attribute Relations

In this activity, we will generate a regression plot to visualize a real-life dataset using the Seaborn library. You have a dataset pertaining to various animals, including their body mass and maximum longevity. To discover whether there is any linear relationship between these two variables, a regression plot will be used.

```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df = pd.read_csv('../../Datasets/anage_data.csv')
df
```

```python
df.columns
```

Filter the data so that you end up with samples containing a body mass and maximum longevity. Only consider samples for the Mammalia class and a body mass of less than 200,000.

```python
data = df[df['Class'] == 'Mammalia']
data = data.loc[:, ['Body mass (g)', 'Maximum longevity (yrs)']]
data = data.dropna(axis=0)
data = data[data['Body mass (g)'] < 200000]
data
```

```python
sns.set()
plt.figure(figsize=(10,8))
sns.regplot(
    x='Body mass (g)',
    y='Maximum longevity (yrs)',
    data=data,
)
plt.show()
```
