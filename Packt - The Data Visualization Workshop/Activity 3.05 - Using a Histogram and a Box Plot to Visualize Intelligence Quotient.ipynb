# Activity 3.05: Using a Histogram and a Box Plot to Visualize Intelligence Quotient

In this activity, we will visualize the intelligence quotient (IQ) of 100 applicants using histogram and box plots. 100 people have come for an interview in a company. To place an individual applicant in the overall group, a histogram and a box plot shall be used.

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
df = pd.read_csv('../../Datasets/iq_scores.csv')
df
```

- Plot a histogram with 10 bins for the given IQ scores. IQ scores are normally distributed with a mean of 100 and a standard deviation of 15.
- Visualize the mean as a vertical solid red line, and the standard deviation using dashed vertical lines. 
- Add labels and a title.

```python
group_b_mean = df.group_b.mean()
group_b_std = df.group_b.std()

fig = plt.figure(figsize=(10,5))
# histogram
plt.hist(df.group_b, bins=10)
# lines
plt.axvline(group_b_mean, color='r', linestyle='-', linewidth=3) # mean
plt.axvline(group_b_mean - group_b_std, color='r', linestyle='--', linewidth=1) # std-lower
plt.axvline(group_b_mean + group_b_std, color='r', linestyle='--', linewidth=1) # std-upper
#labels
plt.xlabel("IQ score")
plt.ylabel("Frequency")
plt.title("IQ scores for a test group of a hundred adults", fontsize=12, fontweight='bold')
plt.show()
```

- Create a box plot to visualize the same IQ scores.
- Add labels and a title.

```python
fig = plt.figure(figsize=(8,5))
# boxplot
plt.boxplot(df.group_b, labels=['Test group'], notch=True)
#labels
# plt.xlabel("")
plt.ylabel("IQ score")
plt.title("IQ scores for a test group of a hundred adults", fontsize=12, fontweight='bold')
plt.show()
```

- Create a box plot for each of the IQ scores of the different test groups.
- Add labels and a title.

```python
fig = plt.figure(figsize=(8,5))
# boxplot
plt.boxplot(df, labels=df.columns, notch=True)
#labels
plt.xlabel("Test groups")
plt.ylabel("IQ score")
plt.title("IQ scores for different test groups", fontsize=12, fontweight='bold')
plt.show()
```
