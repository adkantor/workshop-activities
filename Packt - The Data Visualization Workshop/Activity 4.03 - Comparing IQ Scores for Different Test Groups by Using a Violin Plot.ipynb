# Activity 4.03: Comparing IQ Scores for Different Test Groups by Using a Violin Plot

In this activity, we will compare the IQ scores among four different test groups by using the violin plot that's provided by the Seaborn library. 

```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
# read data
df = pd.read_csv('../../Datasets/iq_scores.csv')
df.columns = ['Group A', 'Group B', 'Group C', 'Group D']
df
```

```python
# reshape data
data = df.melt(var_name='Groups', value_name='IQ score')
data
```

```python
sns.set()
sns.set_style('whitegrid')
sns.set_context('talk')

plt.figure(figsize=(10, 8))
sns.violinplot(
    x='Groups',
    y='IQ score',
    data=data
)
plt.title('IQ scores for different test groups')
sns.despine(left=True, right=True)
plt.show()
```
