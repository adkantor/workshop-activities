# Activity 4.01: Using Heatmaps to Find Patterns in Flight Passengers' Data

In this activity, we will use a heatmap to find patterns in the flight passengers' data. The goal of this activity is to apply your knowledge about color palettes to choose a suitable color palette for this data.

The given dataset contains the monthly figures for flight passengers for the years 1949 to 1960. This dataset originates from the Seaborn library.

```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
# read data
df = pd.read_csv('../../Datasets/flight_details.csv')
df
```

```python
# restructure data
months = [
    'January', 'February', 'March', 
    'April', 'May', 'June', 
    'July', 'August', 'September', 
    'October', 'November', 'December'
]
data = df.pivot(index='Months', columns='Years', values='Passengers')
# data.columns = data.columns.droplevel(0)
data = data.reindex(months)
data
```

```python
# plot heatmap

cmap1 = sns.cubehelix_palette(rot=-0.4, as_cmap=True)
cmap2 = sns.light_palette('DarkCyan', as_cmap=True)

plt.figure(figsize = (12,10))
sns.heatmap(
    data=data, 
    cmap=cmap1
)
plt.title('Flight Passengers from 1949 to 1960')
plt.show()
```
