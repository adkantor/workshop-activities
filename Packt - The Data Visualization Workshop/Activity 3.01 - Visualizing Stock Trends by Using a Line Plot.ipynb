# Activity 3.01: Visualizing Stock Trends by Using a Line Plot

In this activity, we will create a line plot to show stock trends. The aim of this activity is to not just visualize the data but to use labels, a title, and a legend to make the visualization self-explanatory and "complete."

Let's look at the following scenario: you are interested in investing in stocks. You downloaded the stock prices for the "big five": Amazon, Google, Apple, Facebook, and Microsoft. You want to visualize the closing prices in dollars to identify trends. This dataset is available in the Datasets folder that you had downloaded initially. 

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df_amzn = pd.read_csv('../../Datasets/AMZN_data.csv', parse_dates=[0])
df_googl = pd.read_csv('../../Datasets/GOOGL_data.csv', parse_dates=[0])
df_aapl = pd.read_csv('../../Datasets/AAPL_data.csv', parse_dates=[0])
df_fb = pd.read_csv('../../Datasets/FB_data.csv', parse_dates=[0])
df_msft = pd.read_csv('../../Datasets/MSFT_data.csv', parse_dates=[0])
```

```python
fig = plt.figure(figsize=(12,8))
fig.suptitle('Stock prices for the "Big Five"', fontsize=14, fontweight='bold')

plt.plot(df_amzn.date, df_amzn.close, label='Amazon')
plt.plot(df_amzn.date, df_googl.close, label='Google')
plt.plot(df_amzn.date, df_aapl.close, label='Apple')
plt.plot(df_amzn.date, df_fb.close, label='Facebook')
plt.plot(df_amzn.date, df_msft.close, label='Microsoft')

plt.yscale('log') # log scale y axis

plt.xlabel('Date')
plt.ylabel('Close')

plt.legend()
plt.grid()

plt.show()
```
