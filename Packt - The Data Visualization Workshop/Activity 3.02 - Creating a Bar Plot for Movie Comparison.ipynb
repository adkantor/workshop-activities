# Activity 3.02: Creating a Bar Plot for Movie Comparison

In this activity, we will create visually appealing bar plots. We will use a bar plot to compare movie scores. You are given five movies with scores from Rotten Tomatoes. The Tomatometer is the percentage of approved Tomatometer critics who have given a positive review for the movie. The Audience Score is the percentage of users who have given a score of 3.5 or higher out of 5. Compare these two scores among the five movies.

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
import matplotlib.ticker as mtick
```

```python
df = pd.read_csv('../../Datasets/movie_scores.csv', index_col=0)
df
```

```python
df.MovieTitle.values
```

```python
fig = plt.figure(figsize=(12,6))
plt.title("Movie comparison", fontsize=16, fontweight='bold')

labels = df.MovieTitle.values
x = np.arange(len(labels))
width = 0.4
plt.bar(x - width / 2, df.Tomatometer, width=width, label='Tomatometer')
plt.bar(x + width / 2, df.AudienceScore, width=width, label='Audience Score')

# get Axes
ax = plt.gca()
# x axis
ax.set_xticks(x) # add major x tick
ax.set_xticklabels(labels, rotation=20) # add major x tick labels
# y axis
ax.set_yticks(np.arange(0,100,5), minor=True) # add minor y tick
ax.yaxis.set_major_formatter(mtick.PercentFormatter()) # format as percent
ax.yaxis.grid(which='major') # add major horizontal grid
ax.yaxis.grid(which='minor', linestyle='--') # add minor horizontal grid

# legend
plt.legend()

plt.show()
```
